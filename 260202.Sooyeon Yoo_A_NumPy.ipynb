```python
# part2_4 = Create two NumPy arrays: a = np.array([1, 2, 3, 4]) and b = np.array([5, 6, 7, 8]).
```

```python
a = np.array([1, 2, 3, 4])
```

```python
b = np.array([5, 6, 7, 8])
```

```python
# part2_4_1: Add the two arrays element-wise.
```

```python
part2_4_1 = np.empty(4, dtype = int)
```

```python
x1 = 0
```

```python
while x1 < 4:
    part2_4_1[x1] = a[x1] + b[x1]
    x1 += 1
```

```python
print(part2_4_1)
```

```python
# part2_4_2: Multiply the two arrays element-wise.
```

```python
part2_4_2 = np.empty(4, dtype = int)
```

```python
x2 = 0
```

```python
while x2 < 4:
    part2_4_2[x2] = a[x2] * b[x2]
    x2 += 1
```

```python
print(part2_4_2)
```

```python
# part2_4_3: Find the dot product of the two arrays.
```

```python
part2_4_3 = part2_4_2[0] + part2_4_2[1] + part2_4_2[2] + part2_4_2[3]
```

```python
print(part2_4_3)
```

```python
# part2_5_1: Create an array c = np.array([0, np.pi/2, np.pi]). Use NumPy’s trigonometric functions to compute the sine, cosine, and tangent of each element in the array.
```

```python
c = np.array([0, np.pi/2, np.pi])
```

```python
x3 = 0
```

```python
while x3 < 3:
    print(np.sin(c[x3]), np.cos(c[x3]), np.tan(c[x3]))
    x3 += 1
```
